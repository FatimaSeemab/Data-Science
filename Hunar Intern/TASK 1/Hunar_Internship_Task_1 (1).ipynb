## Visualization of data

```python
integer_columns=[column for column in data.columns if (data[column].dtype=="int64" or data[column].dtype=="float64")]
print(integer_columns)
```

```python
integer_column_len=len(integer_columns)
print(integer_column_len)
```

```python
import matplotlib.pyplot as plt
def data_visualization():
  fig, axes = plt.subplots(9,5 ,figsize=(10, 20))  # Create 5 rows of subplots
  axes = axes.flatten()  # Flatten to make it easier to iterate over

  # Initialize the counter
  i = 0

  for column in integer_columns:
      if i >= len(axes) or i>47:
          break  # Ensure we do not try to access more subplots than available
      ax = axes[i]
      i += 1

      ax.boxplot(data[column])
      ax.set_title(column)  # Correct method to set the title

  plt.tight_layout()  # Adjust layout to prevent overlap
  plt.show()
```

```python
data_visualization()
```

## Dealing with Na Values

```python
data.interpolate(method="linear")
```

```python
columns=data.columns
for i in columns:
  print(i, data[i].unique())
```

```python
print()
```

```python
for column in integer_columns:
    data[column]=data[column].fillna(data[column].mean())
```

```python
for column in integer_columns:
    data[column]=data[column].fillna(data[column].mode())
```

```python
for column in integer_columns:
    data[column]=data[column].fillna(data[column].median())
```

## different imputation strategies

```python
data.fillna(method="ffill")
```

```python
data.fillna(method="bfill")
```

## Finding Duplicate Rows

```python
# Identify duplicates
duplicates = data.duplicated()
print("Duplicate Rows:\n", data[duplicates])

# Remove duplicates
df_cleaned = data.drop_duplicates()
```

## Finding Duplicate Columns

```python
def find_duplicate_columns(df):
    duplicate_column_pairs = []
    columns = data.columns
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            if data.iloc[:, i].equals(data.iloc[:, j]):
                duplicate_column_pairs.append((columns[i], columns[j]))
    return duplicate_column_pairs

# Identify duplicate columns
duplicates = find_duplicate_columns(data)
print("Duplicate columns:", duplicates)
```

## Dealing with outliers

```python
import numpy as np
```

```python
for column in integer_columns:
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  data[column] = np.where(data[column] < (Q1 - 1.5 * IQR), Q1 - 1.5 * IQR, data[column])
  data[column] = np.where(data[column] > (Q3 + 1.5 * IQR), Q3 + 1.5 * IQR, data[column])
```
