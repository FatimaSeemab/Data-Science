### Step 4.

```python
df['item_price'] = df['item_price'].apply(lambda x: float(x.strip().split("$")[1]))
```

```python
g=df.groupby("item_name")
for key, data in g:
    # Filter the data based on the conditions
    filtered_data = data[(data['item_price'] > 10) & (data['quantity'] < 2)]
    # Print the item name if the condition is met
    if not filtered_data.empty:
        print(key)
```

```python

for key, data in g:
  data["item_price"]=data["item_price"].apply(lambda x: float(x.strip().split("$")[1]))
  print(len(data["item_price"]))
  for i in range(int(data["item_price"].count())):
    if data.loc["item_price"][i]>10 and data.loc["quantity"][i]<2:
      print(key)
```
